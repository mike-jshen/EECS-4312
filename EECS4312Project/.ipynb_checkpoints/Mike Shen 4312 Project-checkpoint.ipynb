# TASK 2: Preprocess Text

## Import csv files and extract "Review Column"

```python
df1 = pd.read_csv('com.dalthed.tucan.csv')
df2 = pd.read_csv('com.dozingcatsoftware.cameratimer.csv')
df3 = pd.read_csv('fr.tvbarthel.apps.cameracolorpicker.csv')
df4 = pd.read_csv('io.github.sanbeg.flashlight.csv')
df5 = pd.read_csv('wiseguys.radar.csv')
```

```python
desc1 = df1["Review"]
desc2 = df2["Review"]
desc3 = df3["Review"]
desc4 = df4["Review"]
desc5 = df5["Review"]
```

## I. Remove Punctuations

```python
# Remove Punctuations
for (name, data) in desc1.iteritems():
    desc1[name] = desc1[name].translate(str.maketrans('', '', string.punctuation))
    
for (name, data) in desc2.iteritems():
    desc2[name] = desc2[name].translate(str.maketrans('', '', string.punctuation))

for (name, data) in desc3.iteritems():
    desc3[name] = desc3[name].translate(str.maketrans('', '', string.punctuation))
    
for (name, data) in desc4.iteritems():
    desc4[name] = desc4[name].translate(str.maketrans('', '', string.punctuation))

for (name, data) in desc5.iteritems():
    desc5[name] = desc5[name].translate(str.maketrans('', '', string.punctuation))

# Display Output
# for (name, data) in desc2.iteritems():
#     print(data)

# df2[df2.Review.str.contains('Toy Story', case=False)]
```

## II. Remove Special Characters and Emojis

```python
# Remove all special characters using Regrex
for (name, data) in desc1.iteritems():
    desc1[name] = re.sub(r"[^a-zA-Z0-9]",' ',desc1[name])
    
for (name, data) in desc2.iteritems():
    desc2[name] = re.sub(r"[^a-zA-Z0-9]",' ',desc2[name])
    
for (name, data) in desc3.iteritems():
    desc3[name] = re.sub(r"[^a-zA-Z0-9]",' ',desc3[name])
    
for (name, data) in desc4.iteritems():
    desc4[name] = re.sub(r"[^a-zA-Z0-9]",' ',desc4[name])
    
for (name, data) in desc5.iteritems():
    desc5[name] = re.sub(r"[^a-zA-Z0-9]",' ',desc5[name])
```

## III. Numbers to Text

```python
# Convert digit numbers to its corresponding word
for (name, data) in desc1.iteritems():
    desc1[name] = re.sub(r"(\d+)", lambda x: num2words.num2words(int(x.group(0))), desc1[name])
    
for (name, data) in desc2.iteritems():
    desc2[name] = re.sub(r"(\d+)", lambda x: num2words.num2words(int(x.group(0))), desc2[name])
    
for (name, data) in desc3.iteritems():
    desc3[name] = re.sub(r"(\d+)", lambda x: num2words.num2words(int(x.group(0))), desc3[name])

for (name, data) in desc4.iteritems():
    desc4[name] = re.sub(r"(\d+)", lambda x: num2words.num2words(int(x.group(0))), desc4[name])
    
for (name, data) in desc5.iteritems():
    desc5[name] = re.sub(r"(\d+)", lambda x: num2words.num2words(int(x.group(0))), desc5[name])
```

## IV. Remove extra white spaces

```python
# Remove extra Whitespace
for (name, data) in desc1.iteritems():
    # Remove extra white spaces at the start or end
    desc1[name] = desc1[name].strip()
    # Remove extra white space that has been doubled up
    desc1[name] = " ".join(desc1[name].split())
    
for (name, data) in desc2.iteritems():
    # Remove extra white spaces at the start or end
    desc2[name] = desc2[name].strip()
    # Remove extra white space that has been doubled up
    desc2[name] = " ".join(desc2[name].split())
    
for (name, data) in desc3.iteritems():
    # Remove extra white spaces at the start or end
    desc3[name] = desc3[name].strip()
    # Remove extra white space that has been doubled up
    desc3[name] = " ".join(desc3[name].split())
    
for (name, data) in desc4.iteritems():
    # Remove extra white spaces at the start or end
    desc4[name] = desc4[name].strip()
    # Remove extra white space that has been doubled up
    desc4[name] = " ".join(desc4[name].split())
    
for (name, data) in desc5.iteritems():
    # Remove extra white spaces at the start or end
    desc5[name] = desc5[name].strip()
    # Remove extra white space that has been doubled up
    desc5[name] = " ".join(desc5[name].split())
```

## V. Tun all words to lowercase

```python
# Remove uppercase letters
for (name, data) in desc1.iteritems():
    desc1[name] = desc1[name].lower()
    
for (name, data) in desc2.iteritems():
    desc2[name] = desc2[name].lower()
    
for (name, data) in desc3.iteritems():
    desc3[name] = desc3[name].lower()
    
for (name, data) in desc4.iteritems():
    desc4[name] = desc4[name].lower()
    
for (name, data) in desc5.iteritems():
    desc5[name] = desc5[name].lower()
```

## VI. Remove Stop Words

```python
# Remove Stopwords
STOPWORDS = set(stopwords.words('english'))
for (name, data) in desc1.iteritems():
    desc1[name] = ' '.join([word for word in desc1[name].split() if word not in STOPWORDS])

for (name, data) in desc2.iteritems():
    desc2[name] = ' '.join([word for word in desc2[name].split() if word not in STOPWORDS])

for (name, data) in desc3.iteritems():
    desc3[name] = ' '.join([word for word in desc3[name].split() if word not in STOPWORDS])
    
for (name, data) in desc4.iteritems():
    desc4[name] = ' '.join([word for word in desc4[name].split() if word not in STOPWORDS])
    
for (name, data) in desc5.iteritems():
    desc5[name] = ' '.join([word for word in desc5[name].split() if word not in STOPWORDS])
```

##  VII.Lemmatize the Reviews

```python
wordnet_lemmatizer = WordNetLemmatizer()
for (name, data) in desc1.iteritems():
    desc1[name] = wordnet_lemmatizer.lemmatize(desc1[name])

for (name, data) in desc2.iteritems():
    desc2[name] = wordnet_lemmatizer.lemmatize(desc2[name])
    
for (name, data) in desc3.iteritems():
    desc3[name] = wordnet_lemmatizer.lemmatize(desc3[name])
    
for (name, data) in desc4.iteritems():
    desc4[name] = wordnet_lemmatizer.lemmatize(desc4[name])
    
for (name, data) in desc5.iteritems():
    desc5[name] = wordnet_lemmatizer.lemmatize(desc5[name])
```

The choice of stop words should have an impact on the analysis of reviews. Since we are ultimately getting rid of white noise when removing stop words, refining the stop words list to adjust for the specific task of app reviews will further increase accuracy. 

Although the default nltk library has a cover for most of the stop words, there will inevitably be stop words that could be removed in our specific senario or even stop words that should should be kept in the review to maintain the context.

## III. Use Vader for Calculating Review Sentiment

```python
# use vader to get sentiment
analyzer = SentimentIntensityAnalyzer()
vader_polarity1 = []
vader_polarity2 = []
vader_polarity3 = []
vader_polarity4 = []
vader_polarity5 = []

for (name, data) in desc1.iteritems():
    review = analyzer.polarity_scores(data)
    vader_polarity1.append(re.findall(r"[-+]?\d*\.?\d+|\d+", str(review))[-1])
    
for (name, data) in desc2.iteritems():
    review = analyzer.polarity_scores(data)
    vader_polarity2.append(re.findall(r"[-+]?\d*\.?\d+|\d+", str(review))[-1])
    
for (name, data) in desc3.iteritems():
    review = analyzer.polarity_scores(data)
    vader_polarity3.append(re.findall(r"[-+]?\d*\.?\d+|\d+", str(review))[-1])
    
for (name, data) in desc4.iteritems():
    review = analyzer.polarity_scores(data)
    vader_polarity4.append(re.findall(r"[-+]?\d*\.?\d+|\d+", str(review))[-1])
    
for (name, data) in desc5.iteritems():
    review = analyzer.polarity_scores(data)
    vader_polarity5.append(re.findall(r"[-+]?\d*\.?\d+|\d+", str(review))[-1])
    
```

## V. How do sentiments compare with each other? Similarity/Difference? Best option?

The core difference between Textblob and Vader falls under its underlying specification. With Textblob providing a simple API for the user to dive into common language processing. Things suuch as part-speech tagging, noun phrase extraction and common polarity sentiment analysis is done with Textblob. Where as Vader is primarily a lexicon and rule-based sentiment that has a refined dictonary adjusted for social media.

Although the methods seem different, they both extract sentiment polarity within a body of text. Textblob takes the objectivity and the polarity whereas Vader takes the Positive, Negative, Netureal and averages them into a Compounded polarity. Both sentiments range from -1 to 1 where a positive decimal is good sentiment and a negative decimal is bad sentiment.

In my opinion, the best option would be to use the Vader sentiment analysis library. Although Textblob's API is much more refined and easier to use, Vader provides us with a flexible library that is specifically adjusted for internet language and is especially helpful when it comes to app reviews. When looking into Vader's github repository, they recentely expanded upon extracting sentiment from emojis. Using a flexible tool like Vader can provide more accurate results when extracting sentiment from emojis and specific internet language.

## II. Extract 10 LDA Topics w. atleast 7 words.
After preprocessing the Text, we carry out a TFIDF vectorization on the text
### 1. Use TFIDF vectoriztion on text column

This gives us a term matrix which can help us with topic modeling. This vectorization compares the amount of times a word appears in the document with the numbers of documents containing the word.

### 2. Apply LDA on vectorized text.

Use the LDA function to first state the parameters, and then transform the vectorized text to extract requirements.

```python
# Use TFIDF vectoriztion on text column
vect =TfidfVectorizer(stop_words=stop_words,max_features=1000)
vect_text=vect.fit_transform(desc2)

# Apply LDA on vectorized text.
lda_model=LatentDirichletAllocation(n_components=10,learning_method='online',random_state=42,max_iter=1) 
lda_top=lda_model.fit_transform(vect_text)
```

### 3. Check Reaults
We can take a look at the proportion of topics for the CamTimer Application.

```python
print("Document 0: ")
for i,topic in enumerate(lda_top[0]):
    print("Topic ",i+1,": ",topic*100,"%")
```

### 4. Extract and Print topics

Iterate through lda_model and print the associated topics.

```python
vocab = vect.get_feature_names_out()
for i, comp in enumerate(lda_model.components_):
    vocab_comp = zip(vocab, comp)
    sorted_words = sorted(vocab_comp, key= lambda x:x[1], reverse=True)[:10]
    print("Topic "+str(i + 1)+": ")
    for t in sorted_words:
        print(t[0],end=" ")
    print("\n")
```
